```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Load in file
csv_path = "Resources/purchase_data.csv"
```

```python
# Read and display the CSV with Pandas
purchase_file_pd = pd.read_csv(csv_path)
purchase_file_pd.head()
```

```python
total_players = len(purchase_file_pd['SN'].unique())
print(total_players)
```

```python
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

```python
unique_item = len(purchase_file_pd['Item ID'].unique())
average_price = purchase_file_pd['Price'].mean()
total_purchase = purchase_file_pd['Purchase ID'].count()
total_revenue = purchase_file_pd['Price'].sum()
```

```python
print(unique_item)
print(average_price)
print(total_purchase)
print(total_revenue)
```

```python
purchase_analysis_table = pd.DataFrame({"Number of Unique Items": [unique_item],
                                       "Average Price": [average_price],
                                       "Number of Purchase": [total_purchase],
                                       "Total Revenue": [total_revenue]})
purchase_analysis_table
```

```python
purchase_analysis_table["Average Price"] = purchase_analysis_table["Average Price"].map("${:.2f}".format)
purchase_analysis_table["Total Revenue"] = purchase_analysis_table["Total Revenue"].map("${:.2f}".format)
purchase_analysis_table
```

```python
gender_group_table = purchase_file_pd.groupby(['Gender'])
gender_group_table.head()
```

```python
# .nunique helps count distinct value (unique values only)
# .count helps count non-null value
# .size helps count total values including null values
# link https://stackoverflow.com/questions/45759966/counting-unique-values-in-a-column-in-pandas-dataframe-like-in-qlik/45760116
gender_count = gender_group_table["SN"].nunique()
gender_count_pd = pd.DataFrame(gender_count)
gender_count_pd = gender_count_pd.rename(columns={"SN": "Total Count"})
gender_count_pd.head()
```

```python
gender_count_pd = gender_count_pd.sort_values("Total Count", ascending=False)
gender_count_pd.head()
```

```python
gender_percent = gender_count / gender_count_pd["Total Count"].sum() * 100
gender_count_pd["Percentage of Players"] = gender_percent
gender_count_pd.head()
```

```python
gender_count_pd["Percentage of Players"] = gender_count_pd["Percentage of Players"].map("{:.2f}".format)
gender_count_pd.head()
```

```python
gender_purchase_count = gender_group_table["Purchase ID"].count()
gender_purchase_count
```

```python
purchase_by_gender = pd.DataFrame(gender_purchase_count)
purchase_by_gender = purchase_by_gender.rename(columns={"Purchase ID": "Total Purchase"})
purchase_by_gender.head()
```

```python
avg_purchase_price_by_gender = gender_group_table["Price"].mean()
purchase_by_gender["Average Purchase Price"] = avg_purchase_price_by_gender
purchase_by_gender.head()
```

```python
total_purchase_price_by_gender = gender_group_table["Price"].sum()
purchase_by_gender["Total Purchase Value"] = total_purchase_price_by_gender
purchase_by_gender.head()
```

```python
avg_purchase_price_by_gender_per = total_purchase_price_by_gender / gender_count
purchase_by_gender["Avg Total Purchase per Person"] = avg_purchase_price_by_gender_per
purchase_by_gender.head()
```

```python
purchase_by_gender["Average Purchase Price"] = purchase_by_gender["Average Purchase Price"].map("${:.2f}".format)
purchase_by_gender["Total Purchase Value"] = purchase_by_gender["Total Purchase Value"].map("${:.2f}".format)
purchase_by_gender["Avg Total Purchase per Person"] = purchase_by_gender["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_by_gender.head()
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]
```

```python
purchase_file_pd["Age Group"] = pd.cut(purchase_file_pd["Age"], bins, labels=group_labels)
purchase_file_pd.head(10)
```

```python
age_group_table = purchase_file_pd.groupby(['Age Group'])
age_group_table.head(10)
```

```python
age_count = age_group_table["SN"].nunique()
age_count_pd = pd.DataFrame(age_count)
age_count_pd = age_count_pd.rename(columns={"SN": "Total Count"})
age_count_pd
```

```python
avg_purchase_price_by_age = age_group_table["Price"].mean()
age_count_pd["Average Purchase Price"] = avg_purchase_price_by_age
age_count_pd
```

```python
total_purchase_price_by_age = age_group_table["Price"].sum()
age_count_pd["Total Purchase Value"] = total_purchase_price_by_age
age_count_pd
```

```python
average_purchase_price_by_age_per = age_group_table["Price"].sum() / age_count
age_count_pd["Avg Total Purchase per Person"] = average_purchase_price_by_age_per
age_count_pd
```

```python
age_count_pd["Average Purchase Price"] = age_count_pd["Average Purchase Price"].map("${:.2f}".format)
age_count_pd["Total Purchase Value"] = age_count_pd["Total Purchase Value"].map("${:.2f}".format)
age_count_pd["Avg Total Purchase per Person"] = age_count_pd["Avg Total Purchase per Person"].map("${:.2f}".format)
age_count_pd.head()
```
